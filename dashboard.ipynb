```python
from kafka import KafkaConsumer
from json import loads
import uuid
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
myuuid = str(uuid.uuid4())
topic = 'processedPosting'
server = '172.18.0.2:9092'

print(f'''
          ============================================================================
          |                        Configuration [Consumer]                                     |       
          ============================================================================
          | Group id                 {myuuid}
          | Topic                    {topic}                                        
          | Server                   {server}                      
          ============================================================================
     ''')
consumer = KafkaConsumer(
     topic,
     bootstrap_servers=[server],
     auto_offset_reset='latest',
     enable_auto_commit=True,
     group_id='my-group')

userWithNegativePostsMap = {}


for message in consumer:
    msg = message.value.decode('utf-8')
    user, comment = msg.split("::")
    userWithNegativePostsMap[user] = userWithNegativePostsMap.get(user, 0) + 1
#     print(f'{user} ---> {comment}')
    df = pd.DataFrame(data={'users': list(userWithNegativePostsMap.keys()), 'count': list(userWithNegativePostsMap.values()) })
    df2 = df.sort_values(by=['count'], ascending=False).head(5)
    print(list(zip(df2['users'], df2['count'])))
```
